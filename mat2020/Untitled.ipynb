```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def slope_field_7(t, y):
    return (4 - t * y) / (1 + y**2)
def slope_field_8(t, y):
    return - t * y + 0.1 * y**3
```

```python
x_values = np.linspace(-10, 10, 20)
y_values = np.linspace(-10, 10, 20)


fig, axs = plt.subplots(1, 2, figsize=(14, 7))

for x in x_values:
    for y in y_values:
        slope_7 = slope_field_7(x, y)
        axs[0].quiver(x, y, 1, slope_7, headwidth=5, headlength=7)
        slope_8 = slope_field_8(x, y)
        axs[1].quiver(x, y, 1, slope_8, headwidth=5, headlength=7)

axs[0].title.set_text('Direction Field for Equation 7')
axs[1].title.set_text('Direction Field for Equation 8')

for ax in axs:
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    ax.set_xlabel('x')
    ax.set_ylabel("y'")

plt.tight_layout()
plt.show()
```
